```python
from tqdm import tqdm
```

```python
path = 'chromedriver.exe'   # 크롬 드라이버를 사용해서 크롤링
driver = webdriver.Chrome(path)
driver.maximize_window()
driver.get('http://ticket.interpark.com/contents/Ranking/RankList?pKind=C&pCate=RK512&pType=Y&pDate=20210101')
            # 인터파크 인기공연 2019-2021
```

```python
rank_list = driver.find_elements_by_class_name('rankBody')
url_list = []
for rank in rank_list:
    url = rank.find_element_by_css_selector('tr>td>a').get_attribute('onclick')
    url = url.split(",")[0].split("'")[1]
    base_url = 'https://tickets.interpark.com/goods/'
    final_url = base_url + url
    url_list.append(final_url)

print(len(url_list))
print(url_list)
```

```python
#path = 'chromedriver.exe'
#driver = webdriver.Chrome(path)
#driver.maximize_window()

interpark_list = []
for url in tqdm(url_list):
    driver.get(url)
    time.sleep(2)
    
    try:
        title = driver.find_element_by_class_name('prdTitle').text
        place = driver.find_element_by_class_name('infoBtn').text.split('(')[0]

        infoitem_list = driver.find_elements_by_class_name('infoItem')
        for item in infoitem_list:
            infoLabel_list = item.find_elements_by_class_name('infoLabel')
            for label in infoLabel_list:  
                if '공연기간' == label.text:
                    date = item.find_element_by_class_name('infoText').text

                elif '공연시간' == label.text:
                    play = item.find_element_by_class_name('infoText').text

                elif '관람연령' == label.text:
                    age = item.find_element_by_class_name('infoText').text

        gender = driver.find_elements_by_class_name('statGenderValue')
        male = gender[0].text.split('%')[0]
        female = gender[1].text.split('%')[0]

        age_p = driver.find_elements_by_class_name('statAgePercent')
        ten = age_p[0].text.split('%')[0]
        twenty = age_p[1].text.split('%')[0]
        thirty = age_p[2].text.split('%')[0]
        fourty = age_p[3].text.split('%')[0]
        fifty = age_p[4].text.split('%')[0]
    
    except:
        pass

    try:
        data = {'공연명': title, '공연장소': place, '공연기간': date,
           '공연시간': play, '관람연령': age, '남성': male,
           '여성': female, '10대': ten, '20대': twenty, '30대':thirty,
           '40대': fourty, '50대': fifty}
        interpark_list.append(data)
    except:
        pass

print(interpark_list)
```
