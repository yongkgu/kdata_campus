```python
import pandas as pd
```

```python
df = pd.read_excel('./files/최종 상위 50개 지하철역.xlsx')
df
```

```python
pop = pd.read_excel('C:\\Users\\od022\\Desktop\\데청캠 프로젝트 5조\\원시 데이터\\서울 연령대별 인구현황.xlsx')
pop.head()
```

```python
pop = pop[['자치구','10대+20대','40대+50대']]
pop.info()
```

```python
df_pop = pd.merge(df,pop,on= '자치구')
df_pop
```

```python
df_pop.describe()
```

```python
pop['10대+20대'].mean()
```

```python
pop['40대+50대'].mean()
```

```python
list1020 = []
for i in df_pop['10대+20대']:
    if i > pop['10대+20대'].mean():
        list1020.append('O')
    else:
        list1020.append('X')   
```

```python
df_pop['10-20대 평균'] = list1020
df_pop.head()
```

```python
list4050 = []
for i in df_pop['40대+50대']:
    if i > pop['40대+50대'].mean():
        list4050.append('O')
    else:
        list4050.append('X')
    
```

```python
df_pop['40-50대 평균'] = list4050
df_pop.head()
```

```python
df_pop.value_counts('10-20대 평균')
```

```python
df_pop.value_counts('40-50대 평균')
```

```python
all_df = df_pop[(df_pop['10-20대 평균'] == 'O') & (df_pop['40-50대 평균'] == 'O')]  # 1020, 4050 평균이상
all_df.value_counts('자치구') # 최종 지하철역(27개) 키오스크 설치시 모든 공연장르 홍보 최적
```

```python
df_1020 = df_pop[(df_pop['10-20대 평균'] == 'O') & (df_pop['40-50대 평균'] == 'X')]  # 1020만 평균이상
df_1020.value_counts('자치구')  # 최종 지하철역 추천에 포함시키는 역(7개) 키오스크 설치 시 연극, 뮤지컬 장르 홍보 추천
```

```python
df_4050 = df_pop[(df_pop['10-20대 평균'] == 'X') & (df_pop['40-50대 평균'] == 'O')]
df_4050.value_counts('자치구')   # 4050만 평균이상인 지역은 존재 X
```
