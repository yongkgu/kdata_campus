```python
sub_lat = subway['역위도'].values.tolist()
sub_long = subway['역경도'].values.tolist()
per_lat=need['X_CD'].values.tolist()
per_long=need['Y_CD'].values.tolist()
per_num=need['공연건수'].values.tolist()
per_show=need['총예매수(건)'].values.tolist()
```

```python
공연수=[]
공연건=[]
예매건=[]
d=[]
def perr(k):
    per_ct=0
    per_num_cnt=0
    per_show_cnt=0
    for i in range(len(per_lat)):
        SUB=(sub_lat[k],sub_long[k])
        PER=(per_lat[i],per_long[i])
        if haversine(SUB,PER, unit = 'km')<3:
            d.append(haversine(SUB,PER, unit = 'km'))
            per_ct+=1
            per_num_cnt+=per_num[i]
            per_show_cnt+=per_show[i]
            공연수.append(per_ct)
            공연건.append(per_num_cnt)
            예매건.append(per_show_cnt)
            
```
